# **Breast Cancer presiction using classification.**

Importing Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame['label'] = breast_cancer_dataset.target
```

Separating the features and target for performing machine learning

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

**Splitting** the data into training data & Testing data ✂

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

 **Model Training** using logistic regression as we want to do classification that whether it is Benign or Malignant

```python
model = LogisticRegression()
```

Training the Logistic Regression model using training data

```python
model.fit(X_train, Y_train)
```

**Model Evaluation**

using Accuracy Score metric

```python
# accuracy on traiining data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on training data = ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on test data = ', test_data_accuracy)
```

```python
input_data = (13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,0.2699,0.7886,2.058,23.56,0.008462,0.0146,0.02387,0.01315,0.0198,0.0023,15.11,19.26,99.7,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259)

# change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one datapoint
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
  print('The Breast cancer is Malignant')

else:
  print('The Breast Cancer is Benign')
```

```python
# confusion matrix

import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(Y_test, X_test_prediction)

# Create a heatmap of the confusion matrix
sns.heatmap(cm, annot=True, fmt="d")
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.title("Confusion Matrix")
plt.show()
```
